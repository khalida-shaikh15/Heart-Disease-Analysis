```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Heart Disease data.csv")
```

```python
df
```

```python
df.info()
```

```python
bins = [0, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
```

```python
df['sex'] = df['sex'].replace({0: 'female', 1: 'male'})

# Rename the 'sex' column to 'gender
df.rename(columns={'sex': 'gender'}, inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='gender', hue='target',palette='mako')
plt.title('Distribution of Heart Disease by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Heart Disease', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='age_group', hue='target',palette='rocket')
plt.title('Distribution of Heart Disease by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Heart Disease', loc='upper right')
plt.show()
```

```python
sns.barplot(x=df.age_group,y=df.target,hue=df.gender,palette='Paired')
```
